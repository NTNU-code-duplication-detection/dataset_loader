```python
import matplotlib.pyplot as plt

# Data
epochs = [0, 1, 2, 3, 4, 5]
losses = [0.0737, 0.0660, 0.0654, 0.0651, 0.0650, 0.0649]

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(epochs, losses, marker='o', linestyle='-', color='salmon', label='Loss')

# Add titles and labels
plt.title('Training Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)  # Show each epoch on x-axis
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
